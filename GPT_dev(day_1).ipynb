```python
with open('input.txt','r',encoding='utf-8') as file:
  text=file.read()
```

```python
print('length of the dataset :', len(text))
```

```python
chars=sorted(list(set(text)))
vocab_size=len(chars)
print(vocab_size)
```

```python
soti ={ ch:i for i,ch in enumerate(chars)}
itos ={i:ch for i,ch in enumerate(chars)}
encode= lambda s: [soti[c] for c in s]
decode = lambda s: ''.join([itos[c] for c in s])

print(encode("hi hello"))
print(decode(encode("hi hello")))
```

```python
import torch
```

```python
data = torch.tensor(encode(text),dtype=torch.long)
print(data.shape,data.type())
```

```python
n= int(0.9*len(data))
train_data=data[:n]
test_data=data[n:]
```

```python

batch_size=4
block_size=8
def get_batch(string):
  if string=='train':
    data=train_data
  else:
    data=test_data
  ix=torch.randint(len(data)-block_size,(batch_size,))
  x=torch.stack([data[i:i+block_size] for i in ix])
  y=torch.stack([data[i+1:i+1+block_size] for i in ix])
  return x,y

xb,yb=get_batch('train')
print('inputs')
print(xb.shape)
print(xb)
print('targets')
print(yb.shape)
print(yb)
print('============================================================================')
for b in range(batch_size):
  for c in range(block_size):
    context=xb[b,:c+1]
    target=yb[b,c]
    print(f'for this context {context.tolist()} the target is {target}')
```
